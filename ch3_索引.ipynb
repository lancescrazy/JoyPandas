```python
# 复旦大学中体重超过70kg的大四学生，或者北大男生中体重超过80kg的非大四的学生：
condition_1_1 = df_demo.School == 'Fudan University'
condition_1_2 = df_demo.Grade == 'Senior'
condition_1_3 = df_demo.Weight > 70
condition_1 = condition_1_1 & condition_1_2 & condition_1_3
condition_2_1 = df_demo.School == 'Peking University'
condition_2_2 = df_demo.Grade == 'Senior'
condition_2_3 = df_demo.Weight > 80
condition_2 = condition_2_1 & (~condition_2_2) & (condition_2_3)
df_demo.loc[condition_1 | condition_2]
```

### attention: 不要使用链式赋值

在对表或者序列赋值时，应当在使用一层索引器后直接进行赋值操作，

这样做是由于进行多次索引后赋值是赋在临时返回的 copy 副本上的，

而没有真正修改元素从而报出 SettingWithCopyWarning 警告。

例如，下面给出的例子：

```python
df_chain = pd.DataFrame([[0,0],[1,0],[-1,0]], columns=list('AB'))
print(df_chain)
import warnings
with warnings.catch_warnings():
    warnings.filterwarnings('error')
    try:
        df_chain[df_chain.A != 0].B = 1 # 使用方括号列索引后，再使用点的列索引
    except Warning as w:
        Warnings_Msg = w
print(Warnings_Msg)
```

### 5. 随机抽样
> 对DataFrame：每一行看做一个样本，每一列看做一个特征，整个看做整体
>
> sample -- 对样本或特征进行随机抽样
>
> 统计特征来计算数据大致分布，耗费时间
>
> 多统计特征在等概率不放回的随机抽样条件下，是总体统计特征的无偏估计

> 如样本均值，总体均值 -- 从表中抽取部分来做近似估计
>
> ==============================================
>
> sample: n, axis, frac, replace, weights 【抽样数量】【抽样方向】【抽样比例】【是否放回】【抽样相对概率】
>
>

```python
# 以构造的value值相对大小作为抽样概率，做3次有放回抽样
df_sample = pd.DataFrame({'id': list('abcde'),
                          'value': [1, 2, 3, 4, 90]})
print(df_sample)
df_sample.sample(3, replace = True, weights = df_sample.value)
```

### 3. IndexSlice 对象



即使在索引不重复的时候，也能对元祖整体切片，而不能对每层切片，也不允许切片和布尔列表混合使用，引入的IndexSlice就是解决这个问题的。
下面构造一个索引不重复的DataFrame：
> `loc[idex[*, *]]`
>
> `loc[idx[*, *], idx[*, *]]`
>
>

> a. `loc[idx[*, *]]` 不能进行分层分别切片，前一个表示行，后一个是列
>
> b. `loc[idx[*, *], idx[*, *]]` 分层切片， idx前一个行，后一个列
>
>
>
>
>

# 三、索引的常用方法

### 1. 索引层的交换和删除


```python
# 构造一个三级索引的例子，方便理解交换过程
np.random.seed(0)
L1, L2, L3 = ['A', 'B'], ['a', 'b'], ['alpha', 'beta']
mul_index1 = pd.MultiIndex.from_product([L1, L2, L3], 
                                         names = ('Upper', 'Lower', 'Extra'))
L4, L5, L6 = ['C', 'D'], ['c', 'd'], ['cat', 'dog']
mul_index2 = pd.MultiIndex.from_product([L4, L5, L6], 
                                         names = ('Big', 'Small', 'Other'))
df_ex = pd.DataFrame(np.random.randint(-9, 10, (8, 8)),
                     index = mul_index1,
                     columns = mul_index2)
df_ex
```

索引层的交换由 swaplevel （只能交换两个层）和 reorder_levels（指定交换任意层） 完成（ 可以指定交换的是哪一个）

```python
df_ex.swaplevel(0, 2, axis = 1).head()  # 列索引的第一层和第三层交换
```

```python
df_ex.reorder_levels([2, 0, 1], axis = 0).head() # 列表数组代表原来索引的层
```

Attention： 轴之间的索引交换

行或列索引内部的交换，不同方向索引也可交换，后讨论

想要对某个位置元素进行修改，在单层索引时容易出现，先取出索引的 values 属性，再对得到的列表进行修改，最后对index对象重新赋值，。

对多级索引，一个解决方案是把某一层索引临时转为表的元素，然后在修改后，最后定回为索引

map 函数： 定义在index上的方法，传入的不是层的标量值，而是传入索引的元组，方便跨层修改


```python
# 字符串转大写
df_tmp = df_ex.copy()
new_idx = df_tmp.index.map(lambda x:(x[0],
                                     x[1],
                                     str.upper(x[2])))
# 使用函数修改后，对应回到原来的位置
df_tmp.index = new_idx
df_tmp.head()
```

map 的另一个使用方法是多级索引的压缩


```python
df_tmp = df_ex.copy()
new_idx = df_tmp.index.map(lambda x:(x[0] + '-' + 
                                     x[1] + '-' +
                                     x[2]))
# 实现了压缩
df_tmp.index = new_idx
df_tmp.head()  # 单层索引
```

```python
# 对压缩后的df，反向展开
new_idx =df_tmp.index.map(lambda x: tuple(x.split('-')))
df_tmp.index = new_idx
df_tmp.head()
```
